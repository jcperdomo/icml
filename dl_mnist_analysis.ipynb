```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
e1 = "dl_mnist_experiments/deepLearning-multiclass_data_2-untargeted-0.5-1-30-160"
e2 = "dl_mnist_experiments/deepLearning-multiclass_data_2-untargeted-2.2-1-30-160"
e3 = "dl_mnist_experiments/deepLearning-multiclass_data_2-targeted-2.2-1-30-160"
e4 = "dl_mnist_experiments/deepLearning-multiclass_data_2-targeted-0.5-1-30-160"

folder = e2
lh = np.load(folder+"/loss_history.npy")
ma = np.load(folder+"/acc_history.npy")
```

```python
len(ma)
```

```python
num_iters = 20
ensemble_res = .99
plt.plot(range(num_iters), ma, color="orange")
plt.plot(range(num_iters), [ensemble_res] * num_iters, linestyle="--", color="purple")
plt.legend(["untargeted loss", "ensemble baseline"], loc="right", bbox_to_anchor=(.95,.6), fontsize=12)
plt.title("Maximum Accuracy of Neural Network")
plt.ylabel("Max Accuracy")
plt.xlabel("MWU Round")
```
